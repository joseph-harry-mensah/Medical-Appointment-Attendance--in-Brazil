# Data Cleaning

> **Observation**:The column 'ScheduledDay' was a combination of Day and Time; hence I feel the column name should be changed from ScheduledDay to Scheduled_Day_Time instead.Again, AppointmentDay column was in datetime format which should not be so since it is supposed to be a date format. Also, Column AppointmentDay has Datetime instead of date only

```python
# Renaming column ScheduledDay to Scheduled_DayTime

df.rename(columns={'ScheduledDay':'Scheduled_Day_Time',
                   'Hipertension':'Hypertension',
                   'Handcap':'Handicap'})
                
```

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date
```

```python
df
```

### Conclusion from Data Cleaning
> **Result**: Column SheduledDay was renames as Scheduled_Day_Time, Column AppointmentDay was formated to Date instead of Datetime

## Exploratory Data Analysis



### Research Question 1 (How many appiontments did patients "show up" or  "not show up")

```python
#Investigating the count of whether a patient showed up for an appointent
#or a patient did not show up for an appointment 

df['No-show'].value_counts()
```

```python
#Visualizing the value counts

attendance= df['No-show'].value_counts().plot(kind='bar', figsize=(7,7), ylabel= 'Cumulative Frequency', xlabel='Show-up Status')
```


**Observation**: The number of patients who did show up for their appointment day was lesser than the number of patients who did not show up for their appointment.(No-show up =88208, Showed up for appointment =22319)

### Research Question 2 (What age group (Patient's ages) attended their appointment)

```python
# Filtering only records that had "Yes" as show-up


attended = df[df["No-show"] == "Yes"]
attended.head(10)
```

```python
#Filtering only records that had "No" as show-up

not_attended = df[df["No-show"] == "No"]
not_attended.head(10)
```

```python
#Visualizing Ages of patients who showed up for their appointment

attended.Age.value_counts().plot(kind='bar', figsize=(18,10), xlabel='Age',ylabel='Age Count')
```

#### A bar chart visualization the Age counts of patients who showed up for their Appointments

```python
#Visualizing Ages of patients who did not show-up for their appointment


not_attended.Age.value_counts().plot(kind='bar', figsize=(18,10), xlabel='Age',ylabel='Age Count')
```

#### A bar chart visualization the Age counts of patients who did not show up for their Appointments


**Observation**: *1.From the exploration (graph 1), you can notice that there was 0 aged people that had the highest age group of patient attendance. ### Research Question 4 (Did patients who received an SMS attend their appointment)

```python
#Filtering out Patients whon received SMS from Patients who did not receive SMS


receive_sms = attended[attended["SMS_received"] == 1]
receive_sms.head(10)
```

```python
#Investigating to see if patients who received an sms actually showed-up for their appointments


sms_attend= receive_sms[receive_sms["No-show"] == "Yes"]
sms_attend.head(10)
```

```python
#Exploring to see if there were patients who received an sms but did not show-up for their appointments

sms_not_attend= receive_sms[receive_sms["No-show"] == "No"]
sms_not_attend
```

```python
#Investigating to see if patients who did not receive an sms actually showed-up for their appointments

nosms_attend = attended[attended["SMS_received"] == 0]
nosms_attend
```

```python
receive_sms.shape, sms_attend.shape, sms_not_attend.shape, nosms_attend.shape
```

**Observation**: From exploring the dataset, the total numer of entries which patients attended their appointment was 22319 entries. It was discovered that from Patients who received SMS all attended their appointment (a total of 9784 entries of patients who received sms); 
There were patients who also did not receive SMS who also attented their appointments(a total of 12535 entries).

### Research Question 5 (Which Health Condition or Reason was most frequent)

```python
#Investigating the type of appointment which had the highest show-up count

Alcoholism = attended[attended["Alcoholism"] == 1]
Handicap = attended[attended["Handcap"] == 1]
Hypertension = attended[attended["Hipertension"] == 1]
Diabetes = attended[attended["Diabetes"] == 1]
Scholarship = attended[attended["Scholarship"] == 1]
```

```python
#gaining some details (number of columns and rows) of each reason or appointment type

Alcoholism.shape,Hypertension.shape,Handicap.shape,Diabetes.shape,Scholarship.shape
```

```python
#ploting a cumulative bargraph to understand the counts

ax = chart_reasons.plot.bar(x='Reason', y='Cumulative_Frequency', rot=0, figsize=(7,6),
                            ylabel= 'Cumulative_Frequency', xlabel='Reason / Condition')
```

**Observation**: From the graph, it can be noted that the appointment type that was most frequently attended out of the total show-up of appointments was appointments on Hypertention, followed by Scholarship appointment, Diabetes appointment, Alcoholism appointment and Handicap appointments respectively.
